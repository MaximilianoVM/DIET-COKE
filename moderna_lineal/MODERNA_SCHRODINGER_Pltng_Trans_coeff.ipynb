```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir constantes
hbar = 6.582119569e-16 # constante de Planck reducida en eV s

m = 9.10938356e-31     # masa del electrón en kg

a = 1                 # longitud de la barrera

# Definir función del coeficiente de transmisión
def transmision(epsilon, l):
    #epsilon = E / V
    #l = a               # l = a/lambda , lambda = 1
    return 1 / ( 1 + ( np.sinh(l*np.sqrt(1-epsilon)) )**2 / (4*epsilon*(1-epsilon)) )

# Hacer la gráfica de la transmisión como función de la energía para a=l
epsilon = np.linspace(0, 5, 1000)
transmision_al = transmision(epsilon, a)
plt.plot(epsilon, transmision_al)
plt.xlabel('epsilon (E/V)')
plt.ylabel('Transmisión (T)')
plt.title('Transmisión vs E/V')
plt.show()
```

```python
#energia = np.linspace(0, 5, 1000)
#linspace from 0 to 1 with 1000 points


epsilon_li = np.linspace(0, 1, 1000)
longitud_barrera_li = np.linspace(0, 1, 1000)

# matriz vacía
transmisiones = np.zeros((len(epsilon_li), len(longitud_barrera_li)))

# llenar la matriz
for i in range(len(epsilon_li)):
    for j in range(len(longitud_barrera_li)):
        transmisiones[i,j] = transmision(epsilon_li[i], longitud_barrera_li[j])

# Graficar el mapa de color
plt.imshow(transmisiones.T)
plt.colorbar()
plt.xlabel('Energía')
plt.ylabel('Longitud de la barrera')
plt.title('Transmisión vs Energía y Longitud de la barrera ')
plt.show()
```
