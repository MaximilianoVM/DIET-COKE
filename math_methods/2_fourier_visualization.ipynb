```python
import sympy as sym
from sympy.abc import t
from sympy import pi
from sympy.plotting import plot

```

```python
t = sym.symbols('t')
```

```python
x = sym.Piecewise((0, t<0), (1, t<pi/2), (0, t>pi/2))
ser = sym.fourier_series(x, (t, -pi, pi))
```

```python
ser.truncate(3)
```

```python
sym.plot(ser.truncate(3), (t, -pi, pi), xlim=[-4, 4], ylim=[-1, 2])
```

```python

# Define truncate values and colors
truncate_values = [4, 16, 50]
colors = ['blue', 'green', 'purple']

# Create an empty plot
p = sym.plotting.plot(show=False)

# Plot the original function
p.extend(sym.plotting.plot(x, (t, -pi, pi), xlim=[-4, 4], ylim=[-1, 2], line_color='black', label='Original function', show=False))

#sym.plot(ser.truncate(3), (t, -pi, pi), xlim=[-4, 4], ylim=[-1, 2])

# Loop over truncate values
for truncate_value, color in zip(truncate_values, colors):
    # Truncate the series
    truncated_ser = ser.truncate(truncate_value)
    
    # Plot the truncated series
    p.extend(sym.plotting.plot(truncated_ser, (t, -pi, pi), xlim=[-4, 4], ylim=[-1, 2], line_color=color, label=f'Truncate={truncate_value}', show=False))

# Show the legend
p.legend = True

# Show the plot
p.show()
```
