<h2>TRAPEZOIDAL RULE</h2>

```python
# Trapezoidal Method

# Define function to integrate
def C(x):
    return np.cos( (np.pi*x**2)/2 )

def S(x):
    return np.sin( (np.pi*x**2)/2 )

# Implementing trapezoidal method
def trapezoidal(x0,xn,n, func):
    # calculating step size
    h = (xn - x0) / n
    
    # Finding sum 
    if func == 'c' or func == "C":
        integration = C(x0) + C(xn)  
        for i in range(1,n):
            k = x0 + i*h
            integration = integration + 2 * C(k)
            
    elif func == 's' or func == "S":
        integration = S(x0) + S(xn)
        for i in range(1,n):
            k = x0 + i*h
            integration = integration + 2 * S(k)
    else:
        print('You have to choose S or C')
    
    # Finding final integration value
    integration = integration * h/2
    
    return integration
    
```

```python
t = np.linspace(-12, 12, 1000)
```

```python
sss = []
ccc = []
for t_i in t:
    sss.append( trapezoidal(0, t_i, 1000, 'S') )
    ccc.append( trapezoidal(0, t_i, 1000, 'C') )
```

```python
plt.plot(ccc, sss)
#plt.axes().set_aspect("equal")
#plt.show()
```
