```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir constantes
hbar = 6.582119569e-16 # constante de Planck reducida en eV s

m = 9.10938356e-31     # masa del electrón en kg

a = 1                 # longitud de la barrera

# Definir función del coeficiente de transmisión
def transmision(epsilon):
    #epsilon = E / V
    l = a               # l = a/lambda , lambda = 1
    return 1 / ( 1 + ( np.sinh(l*np.sqrt(1-epsilon)) )**2 / (4*epsilon*(1-epsilon)) )

# Hacer la gráfica de la transmisión como función de la energía para a=l
energia = np.linspace(0, 5, 1000)
transmision_al = transmision(energia)
plt.plot(energia, transmision_al)
plt.xlabel('epsilon (E/V)')
plt.ylabel('Transmisión (T)')
plt.title('Transmisión vs E/V')
plt.show()
```

```python
energia = np.linspace(0, 5, 1000)
longitud_barrera = np.linspace(0, 5, 1000)

# matriz vacía
transmisiones = np.zeros((len(energia), len(longitud_barrera)))

# llenar la matriz
for i in range(len(energia)):
    for j in range(len(longitud_barrera)):
        epsilon = energia[i] / longitud_barrera[j]
        transmisiones[i,j] = transmision(epsilon)

# Graficar el mapa de color
plt.imshow(transmisiones, origin='lower')
plt.colorbar()
plt.xlabel('Energía (eV)')
plt.ylabel('Longitud de la barrera (nm)')
plt.title('Transmisión vs Energía y Longitud de la barrera ')
plt.show()
```
